###### a
Notice that we have 2 signs
    and since bi=0,1, we have S can be 1.00,1.01,1.11,1.10 = 4 and
    E can be -1,0,1 = 3 ways
> Hence there are 2 x 4 x 3 = 24 numbers in the system

```python
# b
import matplotlib.pyplot as plt
from numpy import zeros, array, sort

sign= [-1,1]
S= ['00','01','11','10'] 
E= [-1,0,1] 
full = []
for i in sign:
    for e in E:
        for s in S:
            full.append(i*(1+int(s[0])/2+int(s[1])/4)*(2**e))
full = array(full)
z= zeros(full.size)
plt.figure(figsize=(10,5))
plt.grid()
plt.scatter(full,z)
plt.title('Represented on the real line')
plt.show()
print('The numbers in the system from left to right are',sort(full))
```
