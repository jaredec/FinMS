```python
import math
import numpy as np

# Define function
def f(x):
    return math.exp(x)

# Compute the centered difference approximation,
# INPUT: x values, function f, and h values
def centered_difference(x,f,h):
    df=(f(x+h)-f(x-h))/(2*h)
    return df

h1=[(0.1/2**n) for n in range(0,11)] # given h and n
for h in h1:
    print('D=',centered_difference(0.5,f,h))
```

```python
h1=[(0.1/2**n) for n in range(0,11)]  # given h and n
error=[abs(np.exp(0.5)-centered_difference(0.5, f, h)) for h in h1]
for i in range(0,9):
    print((error[i]-error[i+1])/(error[i+1]-error[i+2]))
    
print('')
print('We can verify the quadratic rate of convergence as 2^q = 4, where q = 2')
```

```python
#Need to compute h0
eps= 2**-52 # machine precision
M3=np.exp(1/2)
h0=((3*eps*abs(np.exp(1/2))/M3))**(1/3)

print("The optimal value that gives the minimum total error is h0=", h0)
```

```python
def de(h):
    return ((1/3)*(h)*M3)-((np.exp(1/2)*(eps)/(h**2)))

de(h0)
```

Notice that h0 is of the form c* eps^1/3. Moreover, we can plug into the derivative of the error, getting a number close to 0 so it gives minimum error.

```python
h2=[(1/2**n) for n in range(0,11)]

#Need to comput Richardson's extrapolation of centered difference approximation
def D_h_ext(x, f, h):
    return ((4*centered_difference(x, f, h/2))-centered_difference(x, f, h))/3

richardson=[D_h_ext(0.5, f, h) for h in h2]
richardson

for i in range(0,6):
    print((richardson[i]-richardson[i+1])/(richardson[i+1]-richardson[i+2]))
```

h0 will minimize the error when O(h_0^4)=O(eps/h0), when h = eps^1/5

```python
#computes h_0 for Richardson's extrapolation to get the minimum error
richardson_h_0=(2**-52)**(1/5)
print("The optimal value for the Richardson extrapolation is h_0=",richardson_h_0)
```
