```python
# Question 4

import numpy as np 
from matplotlib import pyplot as plt
```

```python
def fx(x,y):
    xprime = x*(1-y)
    return xprime

def fy(x, y):
    yprime = y*(x-1)
    return yprime

def heun(x0, y0, dt, n):
    x = []
    y = []
    x.append(x0)
    y.append(y0)
    for i in range(1, n):
        x_k1 = fx(x[i-1], y[i-1]) # K1 for x
        y_k1 = fy(x[i-1], y[i-1]) # K1 for y 
        x_k2 = fx(x[i-1] + dt, y[i-1] + dt * x_k1) # K2 for x
        y_k2 = fy(x[i-1] + dt, y[i-1] + dt * y_k1) # K2 for y 
        x.append(x[i-1] + (dt/2)*(x_k1 + x_k2)) # xn+1
        y.append(y[i-1] + (dt/2)*(y_k1 + y_k2)) # yn+1
    return x, y

# Given 
x0_1 = .25
y0_1 = .75

x0_2 = .5
y0_2 = .5

x0_3 = .9
y0_3 = .9

tf  = 8 # Given the interval
n = [2^2, 2^3, 2^4, 2^5, 2^6, 2^7, 2^8, 2^9, 800]
dt = 0.01

# Initial condition a x(0)=.25, y(0)=.75
x_1 = [] 
y_1 = []

# Initial condition b x(0)=.5, y(0)=.5
x_2 = []
y_2 = []

# Initial condition c x(0)=.9 , y(0)=.9
x_3 = []
y_3 = []

for i in range (len(n)):
    # x(0)=.25, y(0)=.75
    x01, y01 = heun(x0_1, y0_1, dt, n[i])
    x_1.append(x01)
    y_1.append(y01)
    
    # x(0)=.5, y(0)=.5
    x02, y02 = heun(x0_2, y0_2, dt, n[i])
    x_2.append(x02)
    y_2.append(y02)
    
    # x(0)=.9 , y(0)=.9
    x03, y03 = heun(x0_3, y0_3, dt, n[i])
    x_3.append(x03)
    y_3.append(y03)
    
for i in range(len(x_1)):
    plt.plot(x_1[i], y_1[i], color = "blue")

plt.title("First initial condition")
plt.show()

for i in range(len(x_2)):
    plt.plot(x_2[i], y_2[i], color= "blue" )
plt.title("Second initial condition")
plt.show()

for i in range(len(x_3)):
    plt.plot(x_3[i], y_3[i], color= "blue")
plt.title("Third initial condition")
plt.show()
```

What do these curves tell you about the behavior of this
prey-predator system?

We can see by the early initial condition plots that there is more prey than predators. The number of predators increase when there is more prey and decrease when there is less prey. Moreover, if we start with a higher time step/step size and move towards a smaller time step/step size, we can see that the system converges.
